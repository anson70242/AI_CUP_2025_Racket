```python
import pandas as pd
```

## Read Data

```python
data_input = pd.read_csv('../../data/test/labels.csv')
data_input
```

```python
test_input = pd.read_csv('../../data/39_Test_Dataset/sample_submission.csv')
test_input
```

## Load Each Prediction

### Gender

```python
gender_df= pd.read_csv('../../predictions/gender_pred_1.csv')
gender_pred = pd.concat([data_input, gender_df], axis=1)
gender_pred
```

```python
cols_to_format = ['label_1', 'label_2']

# Calculate the means
averaged_probs = gender_pred.groupby('unique_id')[cols_to_format].mean().reset_index()

# Create the new DataFrame 'gender_result' with formatted string values
gender_result = averaged_probs.copy() # Start with a copy

for col in cols_to_format:
    gender_result[col] = averaged_probs[col].apply(lambda x: f"{x:.4f}")

# Now 'gender_result' is a DataFrame with the specified columns formatted as strings
print("gender_result DataFrame with formatted strings:")
print(gender_result.head(), "\n", gender_result.shape)
print("\nData types of gender_result:")
print(gender_result.dtypes)
```

### Hand

```python
hand_df= pd.read_csv('../../predictions/hand_pred_1.csv')
hand_pred = pd.concat([data_input, hand_df], axis=1)
hand_pred
```

```python
cols_to_format = ['hand_1', 'hand_2']

# Calculate the means
averaged_probs = hand_pred.groupby('unique_id')[cols_to_format].mean().reset_index()

# Create the new DataFrame 'hand_result' with formatted string values
hand_result = averaged_probs.copy() # Start with a copy

for col in cols_to_format:
    hand_result[col] = averaged_probs[col].apply(lambda x: f"{x:.4f}")

# Now 'hand_result' is a DataFrame with the specified columns formatted as strings
print("hand_result DataFrame with formatted strings:")
print(hand_result.head(), "\n", hand_result.shape)
print("\nData types of hand_result:")
print(hand_result.dtypes)
```

### Year

```python
year_df= pd.read_csv('../../predictions/year_pred_1.csv')
year_pred = pd.concat([data_input, year_df], axis=1)
year_pred
```

```python
cols_to_format = ['year_0', 'year_1', 'year_2']

# Calculate the means
averaged_probs = year_pred.groupby('unique_id')[cols_to_format].mean().reset_index()

# Create the new DataFrame 'year_result' with formatted string values
year_result = averaged_probs.copy() # Start with a copy

for col in cols_to_format:
    year_result[col] = averaged_probs[col].apply(lambda x: f"{x:.4f}")

# Now 'year_result' is a DataFrame with the specified columns formatted as strings
print("year_result DataFrame with formatted strings:")
print(year_result.head(), "\n", year_result.shape)
print("\nData types of year_result:")
print(year_result.dtypes)
```

### Level

```python
level_df= pd.read_csv('../../predictions/level_pred_2.csv')
level_pred = pd.concat([data_input, level_df], axis=1)
level_pred
```

```python
cols_to_format = ['level_2', 'level_3', 'level_4', 'level_5']

# Calculate the means
averaged_probs = level_pred.groupby('unique_id')[cols_to_format].mean().reset_index()

# Create the new DataFrame 'level_result' with formatted string values
level_result = averaged_probs.copy() # Start with a copy

for col in cols_to_format:
    level_result[col] = averaged_probs[col].apply(lambda x: f"{x:.4f}")

# Now 'level_result' is a DataFrame with the specified columns formatted as strings
print("level_result DataFrame with formatted strings:")
print(level_result.head(), "\n", level_result.shape)
print("\nData types of level_result:")
print(level_result.dtypes)
```

## Joint

```python
result_df = pd.merge(gender_result, hand_result, on='unique_id', how='left')
result_df = pd.merge(result_df, year_result, on='unique_id', how='left')
result_df = pd.merge(result_df, level_result, on='unique_id', how='left')
result_df
```

```python
result_df.drop(columns=['label_2', 'hand_2'], inplace=True)
rename_mapping = {
    'label_1': 'gender',
    'hand_1': 'hold racket handed',
    'year_0': 'play years_0',
    'year_1': 'play years_1',
    'year_2': 'play years_2'
}
result_df.rename(columns=rename_mapping, inplace=True)
result_df
```

### Append for 3030.txt and sort

```python
new_row_data_list = [3030, 0.5, 0.5, 0.3333, 0.3333, 0.3333, 0.25, 0.25, 0.25, 0.25]
new_row_dict = {
    'unique_id': new_row_data_list[0],
    'gender': new_row_data_list[1],
    'hold racket handed': new_row_data_list[2],
    'play years_0': new_row_data_list[3],
    'play years_1': new_row_data_list[4],
    'play years_2': new_row_data_list[5],
    'level_2': new_row_data_list[6],
    'level_3': new_row_data_list[7],
    'level_4': new_row_data_list[8],
    'level_5': new_row_data_list[9],
}

new_row_df = pd.DataFrame([new_row_dict])
result_df = pd.concat([result_df, new_row_df], ignore_index=True)
```

```python
result_df.sort_values(by='unique_id', inplace=True)

print("\nFinal DataFrame after sorting by 'unique_id':")
result_df
```
